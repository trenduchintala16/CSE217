**Feature transformations** refer to a set of algorithms that create new features in the data set based on the original set of given data. This is an important step in machine learning, because when we transform a certain set of data, the data may develop more discriminatory power in their new space than when compared to the original one. An example of this would be converting categorical or nominal data types, which have no application in computation, into binary or one-hot encoding, such as turning "positive" and "negative" values into 1 and 0. Another example is scaling and normalizing data, which will be covered in this lab. 

Similarity based algorithms are used when we have features in a dataset that can help to identify which existing class a datapoint belongs to. For example, given a new image of an animal, does the image appear to have more characteristics of a dog, or more characteristics of a cat? We use existing labeled pictures of dogs and cats to train a model, which we can use to classify the new picture based on similarities found between the novel image and the previously developed models of what a cat versus a dog looks like. A commonly used similarity-based prediction algorithm is the k-Nearest-Neighbors, or `kNN` algorithm. The specifics of this algorithm were covered in lecture, and here you will be able to see it applied in action, step by step.

## 2. Feature Transformation and Scaling 

When working with large datasets, our dataset will contain features that vary highly in magnitudes, units, spread, and range. However, many algorithms, such as the `kNN` algorithm we are working with today, use Euclidean distance between two points in their computations. 
> Reminder: $$\|\pmb{p} - \pmb{q} \| = \sqrt{\sum_{j=1}^{d} (p_j - q_j)^2}$$

As you can see from this formula, features with much higher magnitues and spread will weigh in a lot more than features with smaller magnitudes, even though a smaller magnitude doesn't necessarily mean that it's less important! 

<img src="utility/pics/largeVsSmall.png" alt="small vs large"/>

For example, let's assume that one particular set of data has three features. The first feature is a property that ranges from 1-10, the second feature is a property that ranges from 1-100, and the last feature is a property that ranges from 1-10,000 (ie. smaller changes in the larger feature can have disproportionately large effects). When training our classifier with this information, the kNN algorithm would take much more heavily into account the third feature when clustering because of its much more exaggerated difference in values when compared to the first or second features. This may lead to missed classifications. To avoid this, we need to normalize our data. 

> **Caution**: any feature analysis you perform before doing supervised learning has to be done based on the **training data set** _only_! We can't use knowlege about the data distribution coming from the test data, since this would perturb our evaluation results favorably. In other words using the test data for anything else than testing/evaluation would be _cheating_. Some people call this _data snooping_.

### Feature Scaling (Min-Max Scaling)

In order to not have one feature weighted more heavily than another, we must scale this data. Here, we scale features so that they exist within a 0-1 range (sometimes referred to as 'min-max scaling'). This can be done with this formula $$X_{\text{scaled}} = \frac{X - X_\min}{X_\max - X_\min}$$

> **Note** We will need to make sure we use the same transformation on both our training and test sets. That means that we should scale `X_test` by the min and max of `X_train`.

**Try this!** Complete the `min_max_scale` function so that it computes $X_{\text{scaled}}$ and stores it into `X_scaled`. _**Hint** Think about over which values you should take the `min` and `max` over._

```python
def min_max_scale(X_in, min_scale_by=None, max_scale_by=None):
    ''' Computes the min-max scaled transformation of X_in
        * if min_scale_by or max_scale_by is None: compute the min and max based on X_in, then scale X_in
        * if min_scale_by and max_scale_by are proveded: use those to scale X_in. '''
    
    # only return scale parameters if they are computed on X_in
    return_scale_parameters = (min_scale_by is None) or (max_scale_by is None)
        
    # this was adjusted for brevity!
    if min_scale_by is None:
        min_scale_by = X_in.min(axis=0)
    if max_scale_by is None:
        max_scale_by = X_in.max(axis=0)
        
        
    # your code here
    X_scaled = (X_in - min_scale_by)/(max_scale_by - min_scale_by)
    
    
    assert X_scaled.shape == X_in.shape, \
        'The shape of the array X should not change'

    if return_scale_parameters:
        return X_scaled, min_scale_by, max_scale_by
    else: 
        return X_scaled
```

### Standardization

Another way to scale our data is called **standardization**, which is done by computing $$Z = \frac{X - \bar X}{\text{std}(X)},$$ where division is element-wise. Here, $Z$ is the standardized matrix and $\vec s$ is the standard deviation of each value in $X$. For those who are familiar with statistics (don't worry if you're not), this formula computes the [z-scores](https://en.wikipedia.org/wiki/Standard_score) for each value in $X$. The effect of this is to transform the data to have zero mean and unit variance (ie. to be normally distibuted).

> **Note** We will need to make sure we use the same transformation on both our training and test sets. That means that we should scale `X_test` by the mean and standard deviation of `X_train`.

**Try this!** Complete the `standardize` function so that it computes $Z$ and stores it in `Z`. `Hint` Think about over which values you should take the `mean` and `std` over.

```python
def standardize(X_in, mean_scale_by=None, std_scale_by=None):
    ''' Computes the minstandardization of X_in
        * if mean_scale_by or std_scale_by is None: compute the min and max based on X_in, then scale X_in
        * if mean_scale_by and std_scale_by are proveded: use those to scale X_in. '''
    
    return_std_parameters = (mean_scale_by is None) or (std_scale_by is None)
        
    if mean_scale_by is None:
        mean_scale_by = X_in.mean(axis=0)
    if std_scale_by is None:
        std_scale_by = X_in.std(axis=0)
    
    # your code here
    
    Z = (X_in - mean_scale_by)/std_scale_by
    
    
    assert Z.shape == X_in.shape, \
        'The shape of the array X should not change'

    if return_std_parameters:
        assert mean_scale_by is not None, 'mean_scale_by not computed'
        assert std_scale_by is not None, 'std_scale_by not computed'
        return Z, mean_scale_by, std_scale_by
    else: 
        return Z
```

Let's try visualizing the changes made by our scaling procedure with a box plot. In the cell below, we, again, rereate the plot we made earlier, but this time, create box plots for the scaled data.

```python
X_train_std, mean_scale_by, std_scale_by = standardize(X_train)
X_test_std = standardize(X_test, min_scale_by, max_scale_by)

plt.boxplot(X_train_std)
plt.xlabel('Features')
plt.ylabel('Values')
```

### Reviewing kNN
Let's briefly revisit how the `kNN` algorithmn works with a two dimensional dataset, so that we can design one ourselves.

The `kNN` algorithm is known as a "lazy-learner," meaning that it doesn't actually make any computations until we ask the algorithm to make a prediction. After inputting our training data, the algorithm stores the points for each data point and awaits for us to input prediction features before making any computations. 

After loading our `sepal_length` and `sepal_width` data into our `kNN` algorithm, we are ready to start making predictions. When we input our prediction features, our algorithm will calculate the Euclidean distance from each existing datapoint to our input point. The algorithm will then take the `k` nearest neighbors to our input data and calculate the mode, or the label that appears the most frequently within the `k` neighbors. This value is returned as the label of our input features. 

In the graph below, `k = 6`. The points within the dotted circle are the 6 closet points to the star, which is our input point. Would the star be classified as `A` or `B`?

### Implementing and applying `kNN`

We are ready to design our own `kNN` algorithm. We will use our algorithm to predict what kind of species the flower is after inputting the sepal width and length. 

#### Distance Measure

As with $k$-Means, we will also need to have some distance measures here in order to compute the 'similarity' between our target point and those in our dataset. As mentioned earlier, Euclidean distance is a common choice. And, since you have already implemented it, we will provide our implementation for you to use here.

```python
def euclidean(a, b):
    '''Computes the Euclidean distance between point(s) a and b.'''
    
    return np.linalg.norm(a - b, axis=a.ndim > 1)
```

#### Making Predictions

To make predictions, we will need to compute the distance between our target input point and every point in the dataset.

**Try this!** Complete the `compute_distances` function so that it computes the distances between all points in `X` and point `a`. `Hint` Use the multi-point functionality of `euclidean`.

```python
def compute_distances(X, a):
    '''Computes distances between point A and all points in X'''
    
    # your code here
    distances = euclidean(X, a)
    
    assert distances.shape[0] == X.shape[0], 'You should return one distance for each point in X'

    return distances
```

We will also need to find the labels of the $k$ closest points.

In the following cell, we implement the `get_closest_labels` function so that it computes the labels of the $k$ closest points for a _single_ array of distances. We used the [`np.argsort` 🔗](https://www.numpy.org/devdocs/reference/generated/numpy.argsort.html#numpy.argsort) or [`np.argpartition` 🔗](https://www.numpy.org/devdocs/reference/generated/numpy.argpartition.html#numpy.argpartition) to get the indices of values ordered from smallest to largest.

```python
def get_closest_labels(distances, y, k):
    '''
    Retrives the labels of the k points that have the smallest distance
    given DISTANCES and labels Y.
    '''
    
    # adjusted for brevity
    labels = y[np.argpartition(distances, k)[:k]]
    # the idea is to sort (or partially sort in the case shown) the points by distance
    # and then take the top-k closet ones
        
    assert labels.shape[0] == k, 'You should return the labels of the K nearest neighbors'
    
    return labels
```

Let's get the labels of the three closests points to input $(6.2, 3.1)$.

```python
x_star = np.array([[6.2, 3.1]])
distances = compute_distances(X_2D_train, x_star)
get_closest_labels(distances, y_train, 3)
```

### Verify your Implementation 
Here is a quick visualization to verify the correctness of your implementations:

```python
def plt_data(X_2D_train, y_train, X_test, y_test_pred=None):
    for species, code in encoding.items():
        train_idx = y_train == code
        p = plt.scatter(X_2D_train[train_idx,0], X_2D_train[train_idx,1], label=f'{species} ({code})')

        if y_test_pred is not None:
            test_idx = y_test_pred == code
            color = p.get_facecolor()
            plt.scatter(X_test[test_idx,0], X_test[test_idx,1], color=color, marker='*', s=120, label=f'predicted {species} ({code})')
    
    if y_test_pred is None:
        test_idx = np.ones(X_test.shape[0], dtype=bool)
        plt.scatter(X_test[test_idx,0], X_test[test_idx,1], color='gray', marker='*', s=120, label=f'test point')
            

    plt.legend()
    plt.title('X_2D_train, class labels, and test point(s)')
    plt.show()

plt_data(X_2D_train, y_train, x_star)
```

**Try this!** Complete the `predict` function so that it computes the final predicted labels of the points in `X_test` given `X_train`, `y_train` and `k`. Make sure your function works for multiple test points. _**Hint**: you may iterate through each point in `X_test`._

```python
def predict(X_train, y_train, k, X_test):
    '''Computes the final prediction given the closest labels'''
    from statistics import mode
    predictions = []
    # your code here
    for x in X_test:
        distances = compute_distances(X_train, x)
        labels = get_closest_labels(distances, y_train, k)
        predictions.append(mode(labels))
    predictions = np.array(predictions)
    assert predictions.shape[0] == X_test.shape[0], 'You should return the labels of the K nearest neighbors'
    
    return predictions
```

Now, let's test our `predict` function.

```python
y_hat = predict(X_2D_train, y_train, 3, x_star)
print(x_star, y_hat)

y_hat = predict(X_2D_train, y_train, 3, X_2D_test)

# visulaize training and test data
plt_data(X_2D_train, y_train, X_2D_test, y_hat)
```
